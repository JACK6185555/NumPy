```python
# Import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Visualizing distribution of data

### Plotting histogram with function histogram

```python
penguins = sns.load_dataset("penguins", data_home='./data/')
penguins
```

```python
sns.histplot(penguins["bill_length_mm"])
plt.show()
plt.close()
```

```python
sns.histplot(penguins["bill_length_mm"], bins=50)
plt.show()
plt.close()
```

```python
sns.histplot(penguins["bill_length_mm"], bins=50, kde=True)
plt.show()
plt.close()
```

```python
sns.histplot(data=penguins, x="bill_length_mm", bins=50, kde=True, hue='species')
plt.show()
plt.close()
```

```python
sns.histplot(penguins["bill_length_mm"], stat="percent")
plt.show()
plt.close()
```

```python
sns.countplot(data=penguins, x="island")
# countplot() - creates a bar plot where the height of each bar represents the count of observations in each catagory
plt.show()
plt.close()
```

```python
sns.countplot(data=penguins, x="species", hue="island")
# countplot() - creates a bar plot where the height of each bar represents the count of observations in each catagory
plt.show()
plt.close()
```

```python
sns.kdeplot(data=penguins, x='flipper_length_mm')
plt.show()
plt.close()
```

```python
sns.kdeplot(data=penguins, x='flipper_length_mm', hue='species', fill=True)
plt.show()
plt.close()
```

```python
sns.ecdfplot(data=penguins, x='flipper_length_mm')
plt.show()
plt.close()
```

### Plotting univariate histograms

```python
sns.set_theme(style='white')
sns.displot(penguins, x="flipper_length_mm") # create histogram (default of displot() = histogram)
plt.show()
plt.close()
```

```python
# Choosing the bin size
sns.displot(penguins, x="flipper_length_mm", binwidth=10)
plt.show()
plt.close()
```

```python
# Choosing the number of bins
sns.displot(penguins, x="flipper_length_mm", bins=20)
plt.show()
plt.close()
```

```python
# Specifying the bins

bins = range(170,240,20) # binwidth = 20
sns.displot(penguins, x="flipper_length_mm", bins=bins)
plt.show()
plt.close()
```

### Conditioning on other variables

```python
sns.displot(penguins, x="flipper_length_mm", hue="island")
plt.show()
plt.close()
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="island", kind='kde')
plt.show()
plt.close()
```

```python
# Changing visual representation - element = â€˜stepâ€™
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")
plt.show()
plt.close()
```

```python
# Stacked histogram
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple="stack")
plt.show()
plt.close()
```

```python
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple="dodge")
plt.show()
plt.close()
```

```python
# Separate into a number of subplots

sns.displot(penguins, x="flipper_length_mm", row="species") # or col
plt.show()
plt.close()
```

```python
# Separate into a number of subplots

sns.displot(penguins, x="flipper_length_mm", col="species") # or col
plt.show()
plt.close()
```

```python
# common_norm=False, each subset will be normalized independently
# Density normalization scales the bars so that their areas sum to 1
sns.displot(penguins, x="flipper_length_mm", col="species", stat="density", common_norm=False)
plt.show()
plt.close()
```

```python
sns.displot(penguins, x="flipper_length_mm", col="species", stat="probability", common_norm=True)
plt.show()
plt.close()
```

### KDE Plot

```python
sns.displot(penguins, x="flipper_length_mm", kind="kde")
plt.show()
plt.close()
```

```python
# Choosing the smoothing bandwidth 
sns.displot(penguins, x="flipper_length_mm", kind="kde", bw_adjust=0.25) # default bw_adjust=1
plt.show()
plt.close()
```

```python
# Multiple categories KDE
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde")
plt.show()
plt.close()
```

```python
# Stacked KDE
sns.displot(penguins, x="flipper_length_mm", hue="species", multiple="stack", kind="kde")
plt.show()
plt.close()
```

```python
# Stacked KDE
sns.displot(penguins, x="flipper_length_mm", hue="species", kind="kde", fill=True)
plt.show()
plt.close()
```

### Visualizing bivariate distributions

```python
# A bivariate histogram 
g = sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", cmap='Reds', cbar=True)
plt.show()
plt.close()
```

```python
# A bivariate KDE plot
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", kind="kde")
plt.show()
plt.close()
```

```python
# A bivariate hitogram plot with an additional parameter
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", cbar=True)
plt.show()
plt.close()
```

```python
# A bivariate hitogram plot with an additional parameter
sns.displot(penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", kind='kde')
plt.show()
plt.close()
```

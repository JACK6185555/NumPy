```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_excel('Top Games Globally.xlsx')
```

```python
df
```

```python
df=df.drop_duplicates()
df
```

```python
a=df.groupby(['Genre'])['G.S in Thousand'].sum()
all_data=a.sort_values(ascending=False)

all_data
```

```python
genre_counts = df['Genre'].value_counts()
genre_counts
```

```python
plt.bar(all_data.index,all_data.values)
plt.xticks(rotation=90)
plt.title('Games Genre (Users)')
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
text = ' '.join(df['Genre'])
wordcloud = WordCloud(max_font_size=100, max_words=100, background_color="black").generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
es_action=df[df['Genre']=='Action']['E.S in thousand'].sum()
js_action=df[df['Genre']=='Action']['J.S in thousand'].sum()
nas_action=df[df['Genre']=='Action']['N.A.S in thousand'].sum()
os_action=df[df['Genre']=='Action']['O.S in thousand'].sum()

all_data_action=[es_action,js_action,nas_action,os_action]
all_data_action
```

```python
label=['E.S','J.S','N.A.S','O.S']
plt.bar(label,all_data_action)
plt.ylabel('People in thousand',fontsize=12)
plt.title('Number of people in per area(Action Game)')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
#Drow 2 plots
x1 = np.array([0, 1, 2, 3])
y1= np.array([3, 8, 1, 10])

plt.subplot(2,1,1)
plt.plot(x1,y1)

x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])
plt.subplot(2,1,2)
plt.plot(x2,y2)

plt.show()
```

### Word Cloud <i>(Optional)</i>

Package: https://pypi.org/project/wordcloud/ <br>
Documentation: http://amueller.github.io/word_cloud/

Installation: https://anaconda.org/conda-forge/wordcloud

```python
import numpy as np
import pandas as pd
from os import path
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
text = open("data/dummytext.txt", "r").read()
wordcloud = WordCloud().generate(text)
```

```python
wordcloud = WordCloud(max_font_size=100, max_words=100, background_color="black").generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
from PIL import Image
brown_mask = np.array(Image.open("data/brown_mask.png"))
wc = WordCloud(background_color="white", max_words=2000, 
               mask=brown_mask, contour_width=10, contour_color='brown').generate(text)

plt.figure(figsize=[10,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
# store to file
plt.savefig("mask.png", format="png")

plt.figure(figsize=[10,10])
plt.imshow(brown_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis("off")

```

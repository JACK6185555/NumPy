```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('personality_dataset.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
plt.figure(figsize=(12,10))
sns.countplot(x='Personality',data=df)
plt.title('Class Distrubistion of Personality Types')
plt.xlabel('Personality')
plt.ylabel('Count')
plt.show()
```

```python
g=sns.PairGrid(df)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
plt.show()
plt.close()
```

```python
numeric_colums=['Time_spent_Alone', 'Social_event_attendance', 'Going_outside', 'Friends_circle_size', 'Post_frequency']
```

```python

sns.heatmap(df[numeric_colums].corr(),annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data=pd.read_excel('Depressive disorders.xlsx')
data
```

```python
data.rename(columns={'Unnamed: 0': 'Country'}, inplace=True)

data_melted = pd.melt(data, id_vars=['Country'], var_name='Year', value_name='Value')

data_melted['Year'] = data_melted['Year'].astype(int)

specific_countries = ['World', 'North America (WB)', 'Europe & Central Asia (WB)', 'Japan', 'Canada', 'Americas (WHO)', 'European Region (WHO)','Western Pacific (WHO)']
data_filtered = data_melted[(data_melted['Year'] >= 2000) & (data_melted['Year'] <= 2021) & data_melted['Country'].isin(specific_countries)]

plt.figure(figsize=(15, 9))
for country, group in data_filtered.groupby('Country'):
    plt.plot(group['Year'], group['Value'], marker='o', linestyle='-', label=country)
plt.title('Depressive Disorders Values from 2000 to 2021 for Specific Countries')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


specific_countries = ['World']
specific_who = ['Western Pacific (WHO)']

df_world = data_melted[
    (data_melted['Year'].between(2000, 2021)) & 
    (data_melted['Country'].isin(specific_countries))
]

df_who = data_melted[
    (data_melted['Year'].between(2000, 2021)) & 
    (data_melted['Country'].isin(specific_who))
]

combined_df = pd.concat([df_world, df_who])

pivot_df = combined_df.pivot(index='Year', columns='Country', values='Value')


plt.figure(figsize=(12, 6))
bar_plot = pivot_df.plot(kind='bar', width=0.8, colormap='viridis', ax=plt.gca())

plt.title('Depressive Disorders Values from 2000 to 2021', fontsize=14, pad=20)
plt.xlabel('Year', fontsize=12, labelpad=10)
plt.ylabel('Value', fontsize=12, labelpad=10)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.legend(title='Country', bbox_to_anchor=(1.02, 1), borderaxespad=0, fontsize=10)

plt.tight_layout()
plt.show()
```

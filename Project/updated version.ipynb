```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


# Set Seaborn style
sns.set_style("whitegrid")

# Genre sales data (proxy for player preferences)
genre_data = {
    'Genre': ['Action', 'Sports', 'Shooter', 'Role-Playing', 'Platform', 'Misc', 
              'Racing', 'Fighting', 'Simulation', 'Puzzle', 'Adventure', 'Strategy'],
    'North America': [877830000, 683350000, 582600000, 327280000, 447050000, 410240000, 
                     359420000, 223590000, 183310000, 123780000, 105800000, 68700000],
    'Europe': [525000000, 376850000, 313270000, 188060000, 201630000, 215980000, 
              238390000, 101320000, 113380000, 50780000, 64130000, 45340000],
    'Japan': [159950000, 135370000, 38280000, 352310000, 130770000, 107760000, 
             56690000, 87350000, 63700000, 57310000, 52070000, 49460000],
    'Other': [187380000, 134970000, 102690000, 59610000, 51590000, 75320000, 
             77270000, 36680000, 31520000, 12550000, 16810000, 11360000]
}
df_genre = pd.DataFrame(genre_data)

# Prepare data for clustering (regional sales as features)
X = df_genre[['North America', 'Europe', 'Japan', 'Other']].values

# Apply K-means clustering (3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_genre['Cluster'] = kmeans.fit_predict(X)

# Map clusters to labels
cluster_names = {0: 'High Sales', 1: 'Moderate Sales', 2: 'Low Sales'}
df_genre['Cluster Name'] = df_genre['Cluster'].map(cluster_names)

# Visualize clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='North America', y='Europe', hue='Cluster Name', size='Japan', 
                data=df_genre, palette='Set1', sizes=(50, 500))
for i, row in df_genre.iterrows():
    plt.text(row['North America'] + 1000000, row['Europe'], row['Genre'], fontsize=9)
plt.title('Clustering Genres by Sales: Proxy for Player Preferences')
plt.xlabel('North America Sales')
plt.ylabel('Europe Sales')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("whitegrid")

# --- Visualization 1: Viewership Trends (Traditional Sports vs. Esports, 2017-2022) ---
# Sports viewership data (subset to 2017-2022 for overlap with esports)
sports_data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'Basketball': [20380000, 17560000, 15140000, 7450000, 9910000, 12400000],
    'American Football': [111320000, 103470000, 98480000, 101320000, 95200000, 99180000],
    'Football (Soccer)': [13620000, 14630000, 14620000, 11330000, 1800000, 15000000]
}

# Esports viewership data (2017-2022)
esports_data = {
    'Year': [2017, 2018, 2019, 2020, 2021, 2022],
    'League of Legends': [100000000, 75000000, 117000000, 137000000, 149000000, 152000000],
    'Dota 2': [22809450, 23573040, 25783200, 21411120, 21155820, 23927490],
    'Counter-Strike': [20229150, 17058660, 20200500, 30908460, 29759340, 30507450],
    'Mobile Legends': [0, 0, 0, 0, 78519358, 93791477],  # Starts in 2021
    'PUBG Mobile': [36134010, 55603590, 24114990, 16185600, 12644010, 14695530]
}

# Combine for plotting
df_sports = pd.DataFrame(sports_data).melt(id_vars='Year', var_name='Sport', value_name='Viewership')
df_esports = pd.DataFrame(esports_data).melt(id_vars='Year', var_name='Game', value_name='Viewership')
df_sports['Category'] = 'Traditional Sports'
df_esports['Category'] = 'Esports'
df_sports = df_sports.rename(columns={'Sport': 'Name'})
df_esports = df_esports.rename(columns={'Game': 'Name'})
df_combined = pd.concat([df_sports, df_esports])

# Plot
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Viewership', hue='Name', style='Category', markers=True, data=df_combined)
plt.title('Viewership Trends: Traditional Sports vs. Esports (2017-2022)')
plt.xlabel('Year')
plt.ylabel('Viewership')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

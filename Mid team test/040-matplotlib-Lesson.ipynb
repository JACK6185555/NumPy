## Introduction to Plotting

```python
fig = plt.figure(figsize=(3,2))
# Plot data points as lines (default)
plt.plot([1,2,3,4]) # y-values - (0, 1), (1, 2), (2, 3), (3, 4)
plt.ylabel("some numbers")
plt.show() # displaying the figure
```

```python
fig = plt.figure(figsize=(3,2))
# Plot data points as lines (default)
plt.plot([1,2,3,4], color='darkblue') # y-values - (0, 1), (1, 2), (2, 3), (3, 4)
plt.ylabel("some numbers")
plt.show() # displaying the figure
```

```python
fig = plt.figure(figsize=(3,2))
# Plot data points as lines (default)
plt.plot([1,2,3,4], color='k') # y-values - (0, 1), (1, 2), (2, 3), (3, 4)
plt.ylabel("some numbers")
plt.show() # displaying the figure

# color: red (r), blue (b), black (k), green (g)...
```

```python
fig = plt.figure(figsize=(3,2))
# Plot data points as lines (default)
plt.plot([1,2,3,4], color='#12FF56') # y-values - (0, 1), (1, 2), (2, 3), (3, 4)
plt.ylabel("some numbers")
plt.show() # displaying the figure

# color: red (r), blue (b), black (k), green (g)...
```

```python
# Specifying color, marker and line
# Ploting x vs y by providing two list
fig = plt.figure(figsize=(3,2))
plt.plot([0,2,4,6], [1,5,10,15], 'g*--') # green, marker=*, line='--' (dashed)
plt.ylabel("some numbers")
plt.show()
```

```python
# Specifying color, marker and line
# Ploting x vs y by providing two list
fig = plt.figure(figsize=(3,2))
plt.plot([0,2,4,6], [1,5,10,15], 'rD-') # red, marker=D, line='-' (solid) 
plt.ylabel("some numbers")
plt.show()
```

```python
# Specifying color, marker and line
# Ploting x vs y by providing two list
fig = plt.figure(figsize=(3,2))
plt.plot([0,2,4,6], [1,5,10,15], 'bd-.') 
plt.ylabel("some numbers")
# Set the range of x- and y-axis
plt.axis([0,6,0,20]) # [xmin, xmax, ymin ymax]
plt.show()
```

```python
# Ploting multiple data pairs
fig = plt.figure(figsize=(4,3))
plt.plot([2, 4, 6, 8], 'or', [1, 5, 9, 13], 'sg--', [4, 20, 30, 60], '-b>') 
plt.show()
```

```python
# Ploting multiple data pairs
fig = plt.figure(figsize=(4,3))
plt.plot([2, 4, 6, 8], 'or', label='A')
plt.plot([1, 5, 9, 13], 'sg--', label='B')
plt.plot([4, 20, 30, 60], '-b>', label='C') 
plt.legend() # show legend (must use with label)
plt.grid(axis='y')
plt.show()
```

## Basic Text and Legend Functions

```python
plt.figure(figsize=(3,2)) 
plt.plot([0,2,4,6], [1,5,10,15], 'rh--') 
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title of the figure')
plt.show()
```

```python
# Create a Figure object
fig = plt.figure(figsize=(3,2)) 
ax1 = fig.add_subplot(1,1,1)
ax1.plot([0,2,4,6], [1,5,10,15], 'rh--') 
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')
ax1.set_title('Title')
plt.title('Figure title')
plt.show()
```

```python
plt.close()

# Create a Figure object
fig = plt.figure(figsize=(5,3))
# Add two subplots at a 1x2 grid of the Figure object
ax1 = fig.add_subplot(1,2,1) #  1 row, 2 columns 
ax2 = fig.add_subplot(1,2,2)

# Plot on the Axes 'ax'
ax1.plot([0,2,4,6,8,10], [1,5,10,100,23,3], 'yH--') 
ax2.plot([0,2,4,20,8,10], 'gD-') 

# Set labels
ax1.set_xlabel('X Label', fontsize=20)
ax1.set_ylabel('Y Label')
ax2.set_xlabel('ax2 X Label')
ax2.set_ylabel('ax2 Y Label')

# Set Limits
ax1.set_xlim(0, 10)
ax1.set_ylim(0, 150)
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 150)

# Set a figure title
fig.suptitle('Suptitle', fontsize=30, fontweight='normal')
# font weight: 'light', 'normal', 'bold', 'semibold', 'heavy'
# font weight can be an integer

# Set axes title
ax1.set_title('ax1 title', fontsize=20, fontweight='bold')
ax2.set_title('ax2 title', fontsize=10)

plt.tight_layout() # readjust the plot position (minimize overlapping)
plt.show()
```

```python
# Create a Figure object
fig = plt.figure(figsize=(3,2))
# Add one subplots at a 1x1 grid of the Figure object
ax1 = fig.add_subplot(1,1,1)

ax1.plot([5],[5],'ro') # draw a red dot at coordinate (5, 5)
ax1.set_xlim(0,10) 
ax1.set_ylim(0,20)

# Add the text 'Example of Annotate' at the coordinate (2, 17)
# ha (horizontal alignment): center, left, right
# va (vertical alignment): center, top, bottom
ax1.annotate('Example of Annotate', xy=(2,17), ha='left', va='center', bbox={'facecolor':'green', 'alpha':0.2, 
                                                                             'pad':20})

# Add the text 'Pointing at a data point' at the coordinate (8, 10)
# point to a data point at (5,5)
ax1.annotate('Pointing at a data point', xy=(5,5), xytext=(8,10), ha='right', 
             arrowprops=dict(facecolor='g', shrink=0.1))

plt.show()
```

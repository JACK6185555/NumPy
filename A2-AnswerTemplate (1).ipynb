### Task 2: Data Processing and Visualization

#### 1. Visualize the correlation between the nutrients. 


```python
import seaborn as sns

corr_matrix=df.select_dtypes(include=['number']).corr()
```

```python
sns.heatmap(corr_matrix, cmap="coolwarm")
plt.title("Correlation Matrix of Nutrients")
plt.show()
```

#### 2. Visualize the nutritional content of a croissant. 

```python
croissant = df[df['name'] == 'croissant'].select_dtypes(include=['number'])


croissant_t = croissant.T
croissant_t.columns = ['value']
croissant_sorted = croissant_t.sort_values(by='value', ascending=False)
colors = ['#ff9999' if i == 0 else '#cce5ff' for i in range(len(croissant_sorted))]
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=croissant_sorted.index, y=croissant_sorted['value'], palette=colors)
plt.xticks(rotation=90, ha='right',fontsize=7.4)
plt.title("Nutritional Content of a Croissant")
plt.xlabel('The nutritional content per')
plt.ylabel("Amount per Serving")
```

#### 3. Visualize the Vitamin A content across different food items. 

```python
vitamin_a_sorted = df.sort_values('Vitamin A (IU)', ascending=False)
```

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='Vitamin A (IU)', y='name', data=vitamin_a_sorted, palette='viridis')

plt.title('Vitamin A content of each food item (IU/g)', fontsize=16)
plt.xlabel('Vitamin A (IU)', fontsize=12)
plt.ylabel('Food Item', fontsize=12)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
sns.set_style("whitegrid")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df=pd.read_excel('Book1 (3).xlsx',sheet_name=['Population','Death Rate','Crime Rate & Statistics','Youth Unemployment Rate','Poverty Rate','GDP'])
df_population=df['Population']
df_gdp=df['GDP']
```

```python
df_population
```

```python
df_population
population_df = df_population.loc[:, ~df_population.columns.str.contains("Unnamed")]
country_columns = ['United States', 'China', 'Japan', 'South korea', 'Taiwan', 
                   'Australia', 'Brazil', 'South Africa', 'Italy', 'Spain']
population_df = df_population[country_columns]
population_df['Year'] = df_population['Unnamed: 0']
population_df = population_df.set_index('Year')

population_df
```

```python
df_gdp
gdp_df = df_gdp.loc[:, ~df_gdp.columns.str.contains("Unnamed")]
country_columns = ['United States', 'China', 'Japan', 'South korea', 'Taiwan', 
                   'Australia', 'Brazil', 'South Africa', 'Italy', 'Spain']
gdp_df_df = df_gdp[country_columns]
gdp_df['Year'] = df_gdp['Unnamed: 0']
gdp_df = gdp_df.set_index('Year')
gdp_df
```

```python
countries = ['United States', 'China', 'Japan', 'South korea', 'Taiwan',
             'Australia', 'Brazil', 'South Africa', 'Italy', 'Spain']



plt.figure(figsize=(14, 8))
for country in countries:
    plt.plot(population_df.index, population_df[country], label=country)

plt.title('Population Trends (1950–2024)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
gdp_df = gdp_df.apply(lambda col: col.replace('[\$,B]', '', regex=True).astype(float))

countries_to_plot = ['United States', 'China', 'Japan', 'South Korea', 'Taiwan',
                     'Australia', 'Brazil', 'South Africa', 'Italy', 'Spain']

rename_map = {
    'south korea': 'South Korea',
    'South korea': 'South Korea',
    'south Africa': 'South Africa',
    'South africa': 'South Africa'
}
gdp_df.rename(columns=rename_map, inplace=True)

gdp_selected = gdp_df[countries_to_plot]

plt.figure(figsize=(14, 8))
for country in countries_to_plot:
    plt.plot(gdp_selected.index, gdp_selected[country], label=country)

plt.title("GDP Over Time (in Billions USD)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("GDP (Billions of USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

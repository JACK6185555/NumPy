## Basic Plots

### Bar Chart

```python
#  plt.bar(catgories (list), values (list))
plt.bar(['A', 'B', 'C', 'D'], [20, 25, 40, 10])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Chart')

plt.show()

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar
```

```python
#  plt.bar(catgories (list), values (list))
plt.bar(['A', 'B', 'C', 'D'], [20, 25, 40, 10], [0.5, 0.5, 1, 1])
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Chart')

plt.show()

# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar
```

```python
import numpy as np

labels = ['A', 'B', 'C', 'D'] # [0, 1, 2, 3]
x = np.arange(len(labels))
width = 0.4

# Logical error - two set of bars overlapped 
# plt.bar(x, [20, 25, 40, 10], width=width, label='Data 1') 
# plt.bar(x, [30, 15, 30, 20], width=width, label='Data 2')

# Grouped bar chart 
# Manually position the bars
plt.bar(x - width/2, [20, 25, 40, 10], width=width, label='Data 1') 
plt.bar(x + width/2, [30, 15, 30, 20], width=width, label='Data 2')
# x position = [0, 1, 2, 3]
# x-width/2 position (shift left) = [-0.5, 0.5, 1.5, 2.5]
# x+width/2 position (shift right) = [0.5, 1.5, 2.5, 3.5]


#plt.legend()

# Get the current Axes
ax = plt.gca() 

ax.legend()

# Set the xticks to be x
ax.set_xticks(x) # [-0.5, 0, 0.5, ..., 3.5] -> [0, 1, 2, 3]
# Set the xticklabels
ax.set_xticklabels(labels) # [0, 1, 2, 3] -> ['A', 'B', 'C', 'D']

plt.show()
```

### Stacked Bar Chart

```python
import matplotlib.pyplot as plt

labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
width = 1/len(x)       # the width of the bars

fig, ax = plt.subplots() # Create 1 Axes

# Logical errors - bars are overlapped 
#ax.bar(labels, men_means, width, label='Men')
#ax.bar(labels, women_means, width, label='Women')

ax.bar(labels, men_means, width, label='Men')
ax.bar(labels, women_means, width, bottom=men_means,label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
```

```python
plt.pie([0.4, 0.3, 0.2, 0.1], explode=(1, 0.2, 0.3, 0.5), 
        labels=['A', 'B', 'C', 'D'], autopct='%1.2f%%', pctdistance=0.4)
# %: start of formatter 
# 1.1f minimum width = 1, .1 = 1 dp, f = floating point 
# %%: indicates the symbol %

# "%1.2f" -> 40.00, 30.00, etc
# "%1.2f%%"
plt.show()
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html#matplotlib.pyplot.pie
```

```python
plt.pie([0.4, 0.3, 0.2, 0.1], explode=(0, 0.5, 0, 0), 
        labels=['A', 'B', 'C', 'D'], 
        colors=['pink', 'lightgrey', 'blue', 'yellow'], autopct='%1.2f%%',
        shadow=True, labeldistance=0.8, startangle=-90, radius=1.1, 
        counterclock=False)
plt.show()
```

```python
# runtime error if normalize=False & sum of values != 1
plt.pie([1, 2, 3], explode=(0, 0.1, 0, 0), 
        labels=['A', 'B', 'C', 'D'], 
        colors=['pink', 'lightgrey', 'blue', 'yellow'], autopct='%1.2f%%',
        wedgeprops={'width':1, 'edgecolor':'red'}, normalize=False)
plt.show()
```

### Histogram

```python
import random

# generate a list of 100 random numbers with range from 1 to 50 (inclusive)
x = [random.randint(1,50) for x in range(1000)]

plt.hist(x, bins=20, density=False)
plt.xlim(1,50)
plt.grid()
plt.show()
```

```python
x = [random.randint(1,50) for x in range(1000)]

plt.hist(x, bins=50, density=False)
plt.xlim(1,50)
plt.grid()

mean_value = np.mean(x)
mean_value

# add a vertical line to indicate the mean 
plt.axvline(x=mean_value, color='r', linestyle= '--')
# plt.axvline(x=10, color='r', linestyle= '--')


plt.show()
```

```python
x = [random.randint(1,50) for x in range(1000)]

plt.hist(x, bins=50, density=False, range=(10, 20))
plt.xlim(1,50)
plt.grid()

mean_value = np.mean(x)
mean_value

# add a vertical line to indicate the mean 
plt.axvline(x=mean_value, color='r', linestyle= '--')

plt.show()
```

### Box Plot

```python
import random
import numpy as np

# generate some random test data
all_data = [np.random.normal(0, std, size=100) for std in range(1, 4)]
labels = ['x1', 'x2', 'x3']

plt.boxplot(all_data, vert=True,  # vertical box alignment
            patch_artist=True,    # fill with color
            showfliers=True,
            showmeans=True,
            notch=True, 
            boxprops=dict(facecolor='pink'),
           labels=labels) 
plt.show()
```

### Scatter Plot

```python
import numpy as np

N = 50

# np.random.rand(N): Return a sample (or samples) from the “standard normal” distribution
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```

### Bubble Plot

```python
N = 50

# np.random.rand(N): Return a sample (or samples) from the “standard normal” distribution
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.colorbar()
plt.show()
```

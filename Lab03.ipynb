### Task 1

<p>Visualize the average values of 'area' of different classes using a bar chart according to the following specifications:</p>
<ul>
<li>x-axis: Bean classes</li>
<li>y-axis: Area</li>
<li>x-tick labels: Rotate 45 degrees</li>    
<li>title: Average Area by Bean Class</li>
<li>bar width: 0.5</li>
<li>bar color: orange</li>
</ul>


```python
### Your code here
# Step 1:Load dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df= pd.read_excel("Dry_Bean_Dataset.xlsx")

```

```python
# Step 2: Get the values of 'area'
average_area = df.groupby('Class')['Area'].mean()

width=0.5
```

```python
#Step 3: Plot the graph
plt.bar(average_area.index,average_area,width=width,color='orange',)

plt.xticks(rotation=45)

ax=plt.gca()

ax.set_title('Average Area by Bean Class')

plt.show()
```

### Task 2

<p>Visualizes display the distribution of 'Compactness' of the each class using a horizontal box plot according to the following specification:</p>
<ul>
    <li>The data for each class should be included in the box plot.</li>
    <li>x-axis: Class</li>        
    <li>y-axis: Compactness</li>
    <li>title: Compactness distribution</li>    
    <li>x-ticks: Name of the classes</li>
    <li>box border color: blue</li>
    <li>box color: pink</li>
    <li>Outliers: True</li>
    <li>Show median: True</li>
    <li>Show mean: False</li>
</ul>
</p>


```python
### Your code here

#Step 1:Get the values
x=df['Class']
labels=set(x.values)
barbunya=df[df['Class'] == 'BARBUNYA']['Compactness'].values
bombay=df[df['Class']=='BOMBAY']['Compactness'].values
cail=df[df['Class']=='CALI']['Compactness'].values
dermason=df[df['Class']=='DERMASON']['Compactness'].values
seker=df[df['Class']=='SEKER']['Compactness'].values
sira=df[df['Class']=='SIRA']['Compactness'].values
horoz=df[df['Class']=='HOROZ']['Compactness'].values
all_data=[sira,cail,seker,bombay,horoz,barbunya,dermason]

```

```python
#Step 2: Plot the graph

plt.boxplot(all_data,vert=True,
            patch_artist=True,   
            showfliers=True,
            showmeans=False,
            notch=True, 
            boxprops=dict(edgecolor='blue',facecolor='pink'),
           labels=labels) 
plt.xticks(rotation=45)
plt.show()
```

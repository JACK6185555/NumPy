## Figure and Axes

```python
# Create a figure
fig = plt.figure(figsize=(10,2))

# Create some subplots
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

# Generate the content of each subplot
ax1.plot([1,2,3,4])
ax2.plot([2,4,3,7])
ax3.plot([10,2,5,8,9,10,2,4,2,5,6,8,9])

# Display the figure
plt.show()
```

```python
fig = plt.figure() # empty figure with no subplots
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
axs = fig.subplots(nrows=3,ncols=1)
axs[0].plot([1,2,4])
axs[1].plot([4,5,2,8])
axs[2].plot([1,1,2,3,4,8,1,32,10])
plt.show()
```

```python
# Close all the figures to release resources
plt.close()
```

## Introduction to Plotting

```python
# Plot data points as lines (default)
plt.plot([1,2,3,4]) 
plt.ylabel("some numbers")
plt.show() # displaying the figure
```

```python
# Plot data points as points
plt.plot([1,2,3,4], 'o')
plt.ylabel("some numbers")
plt.show()
```

```python
# Specifying color, marker and line
# Ploting x vs y by providing two list
plt.plot([0,2,4,6], [1,5,10,15], 'g*--') 
plt.ylabel("some numbers")
plt.show()
```

```python
# Specifying color, marker and line
# Ploting x vs y by providing two list
plt.plot([0,2,4,6], [1,5,10,15], 'bd--') 
plt.ylabel("some numbers")
# Set the range of x- and y-axis
plt.axis([0,6,0,20]) # [xmin, xmax, ymin ymax]
plt.show()
```

```python
# Ploting multiple data pairs
plt.plot([2, 4, 6, 8], 'or', [1, 5, 9, 13], 'sg--', [4, 20, 30, 60], '-b>') 
plt.show()
```

## Save Figure

```python
# Create a Figure object
plt.figure() 
# Plot
plt.plot([1,2,3,4,5],[2,3,4,5,10], '*c--')
# Save as a file named 'lineplot.png'
plt.savefig('lineplot.png', dpi=300, bbox_inches='tight') #bbox_inches='tight' removes outer write margins
```

## Basic Text and Legend Functions

```python
# Plot on the default canvas
plt.plot([0,2,4,6], [1,5,10,15], 'rh--') 

# Set labels
plt.xlabel('X Label')
plt.ylabel('Y Label')
```

```python
# Create a Figure object
fig = plt.figure()

# Add two subplots at a 1x2 grid of the Figure object
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

# Plot on the Axes 'ax'
ax1.plot([0,2,4,6,8,10], [1,5,10,100,23,3], 'yH--') 
ax2.plot([0,2,4,20,8,10], 'gD-') 

# Set labels
ax1.set_xlabel('X Label')
ax1.set_ylabel('Y Label')

# Set Limits
ax1.set_xlim(0,10)
ax1.set_ylim(0,150)

# Set a figure title
fig.suptitle('Suptitle', fontsize=10, fontweight='bold')

# Set axes title
ax1.set_title('ax1 title')
ax2.set_title('ax2 title')

plt.show()
```

```python
# Create a Figure object
fig = plt.figure()

# Add one subplots at a 1x1 grid of the Figure object
ax1 = fig.add_subplot(1,1,1)

ax1.plot([5],[5],'ro')
ax1.set_xlim(0,10)
ax1.set_ylim(0,20)

# Add the text 'Example of Annotate' at the coordinate (2, 17)
# ha (horizontal alignment): center, left, right
# va (vertical alignment): center, top, bottom
ax1.annotate('Example of Annotate', xy=(2,17), ha='left', va='center', bbox={'facecolor':'yellow', 'alpha':0.5, 'pad':10})

# Add the text 'Pointing at a data point' at the coordinate (8, 10)
# point to a data point at (5,5)
ax1.annotate('Pointing at a data point', xy=(5,5), xytext=(8,10), ha='right', 
             arrowprops=dict(facecolor='b', shrink=0.05))

plt.show()
```

```python
plt.plot([1, 2, 3], label='Label 1') 
plt.plot([2, 4, 3], label='Label 2') 
plt.legend()
plt.grid()
plt.show()
```
